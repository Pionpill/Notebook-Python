# matplotlib 库基础

## 简明 matplotlib API 入门

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = np.arange(10)
plt.plot(data)
````

### 图片与子图

````python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
````

````python
ax3.plot(np.random.randn(50).cumsum(),'k--')
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*np.random.randn(30))
fig
````

#### pyplot.subplots
- 方法原型
  ```py
  subplots(self, nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None)
  ```

#### .subplots_adjust()
- 方法原型
  ```py
  subplots_adjust(self, left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
  ```

````python
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
````

### 颜色，标记和线类型

````python
data = np.random.randn(30)
plt.plot(data,'ko--')
````

````python
plt.plot(data,color='g',linestyle='dashed',marker='o')
````

````python
plt.plot(data,color='k',linestyle='dashed',marker='o',label='default')
plt.plot(data,color='k',drawstyle='steps-post',label='step-posy')
plt.legend(loc='best')
````

### 刻度，标签和图例

````python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
````

````python
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
fig
````

````python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(100).cumsum(),'k',label='one')
ax.plot(np.random.randn(100).cumsum(),'k--',label='two')
ax.plot(np.random.randn(100).cumsum(),'k.',label='three')
ax.legend(loc='best')
````

### 注释与子图加工

````python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
circle = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
ax.add_patch(circle)
````
